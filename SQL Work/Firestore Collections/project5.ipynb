### Create Apps Collection

#### The key benefits and reviews subcollections had some characters (".", "/") in their title & author fields, respectively, that prevented firestore from creating a proper id. We had no way to bypass this error, so we decided to use default indexing provided by firestore as the id for these two subcollections and just included the two other fields as part of the subcollection.

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

csv_path = "fs_shopify/apps.csv"

df = pd.read_csv("fs_shopify/apps.csv", sep=',', header=0, lineterminator='\n')
app_rows = df.values.tolist()

for a_row in app_rows:
    record = {}
    record['id'] = a_row[0]
    record['url'] = a_row[1]
    record['title'] = a_row[2]
    record['developer'] = a_row[3]
    record['developer_link'] = a_row[4]
    record['icon'] = a_row[5]
    record['rating'] = a_row[6]
    record['review_count'] = a_row[7]
    
    app_ref = db.collection('Apps').document(a_row[0])
    batch.set(app_ref, record)
    

batch.commit()

```

#### Add Reviews to Apps as a subcollection

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()


def batch_commit(rows):
    for row in rows:
        rev_record = {}
        
        rev_record['author'] = (row[1])
        rev_record['rating'] = row[2]
        rev_record['posted_at'] = row[3]
        
        review_ref = db.collection('Apps').document(row[0]).collection('Review').document()
        batch.set(review_ref, rev_record)
    batch.commit()

csv_path = "fs_shopify/reviews.csv"
df = pd.read_csv(csv_path, sep=',', header=0, lineterminator='\n')
rows = df.values.tolist()


chunk_size = 500
for i in range(0, len(rows), chunk_size):
    chunk = rows[i:i+chunk_size]
    batch_commit(chunk)
```

#### Add Key_Benefits to Apps as a subcollection

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

def batch_commit(rows):
    for row in rows:
        record = {}
        record['title'] = row[1]
        record['description'] = row[2]
    
        kb_ref = db.collection('Apps').document(row[0]).collection('Key_Benefits').document()
        batch.set(kb_ref, record)
    batch.commit()

csv_path = "fs_shopify/key_benefits.csv"
df = pd.read_csv(csv_path, sep=',', header=0, lineterminator='\n')
rows = df.values.tolist()

chunk_size = 500
for i in range(0, len(rows), chunk_size):
    chunk = rows[i:i+chunk_size]
    batch_commit(chunk)
```

#### Add Pricing_Plans to Apps as a subcollection

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

def batch_commit(rows):
    for row in rows:
        record = {}
        record['id'] = row[0]
        record['title'] = row[2]
        record['price'] = row[3]
    
        pp_ref = db.collection('Apps').document(row[1]).collection('Pricing_Plans').document(row[0])
        batch.set(pp_ref, record)
    batch.commit()

csv_path = "fs_shopify/pricing_plans.csv"
df = pd.read_csv(csv_path, sep=',', header=0, lineterminator='\n')
rows = df.values.tolist()

chunk_size = 500
for i in range(0, len(rows), chunk_size):
    chunk = rows[i:i+chunk_size]
    batch_commit(chunk)
```

### Create Category 

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df_cat = pd.read_csv("fs_shopify/categories.csv", sep=',', header=0, lineterminator='\n')
rows1 = df_cat.values.tolist()

for row in rows1:
    cat_record = {}
    cat_record['id'] = row[0]
    cat_record['title'] = row[1]

    cat_ref = db.collection('Category').document(row[0])
    batch.set(cat_ref, cat_record)
    
batch.commit()
```

#### Add Apps to Category as a subcollection

```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()
batch = db.batch()

df_ac = pd.read_csv("fs_shopify/apps_categories.csv", sep=',', header=0, lineterminator='\n')
rows_ac = df_ac.values.tolist()

df_app = pd.read_csv("fs_shopify/apps.csv", sep=',', header=0, lineterminator='\n')
rows_app = df_app.values.tolist()

for row_ac in rows_ac:
    for a_row in rows_app:
        if row_ac[0] == a_row[0]:
            cat_record = {}
            cat_record['id'] = a_row[0]
            cat_record['url'] = a_row[1]
            cat_record['title'] = a_row[2]
            cat_record['developer'] = a_row[3]
            cat_record['developer_link'] = a_row[4]
            cat_record['icon'] = a_row[5]
            cat_record['rating'] = a_row[6]
            cat_record['review_count'] = a_row[7]
            
            a_ref = db.collection('Category').document(row_ac[1]).collection('Apps').document(a_row[0])
            batch.set(a_ref, cat_record)
batch.commit()
            
```

### Queries

#### Document Count in Collection/Subcollections

##### Since we made Review, Pricing Plans, and Key Benefits subcollections of each app document in the Apps top-level collection, we have to print the subcollection count of each app value (more than 3000 values). This gives a very large result

```python
from google.cloud import firestore
from google.cloud.firestore_v1.base_query import FieldFilter
from google.cloud.firestore_v1 import aggregation

db = firestore.Client()
app_ref = db.collection('Apps')
query = app_ref.count()
result = query.get()
print(f'Apps collection has: {result[0][0].value} documents')

all_apps = app_ref.stream()

for app in all_apps:
    app = app.to_dict()
    pp_ref = app_ref.document(app['id']).collection('Pricing_Plans').count()
    pp_result = pp_ref.get()
    r_ref =  app_ref.document(app['id']).collection('Review').count()
    r_result = r_ref.get()
    kb_ref = app_ref.document(app['id']).collection('Key_Benefits').count()
    kb_result = kb_ref.get()
    
    print(f"Key Benefits subcollection count for App Document {app['id']} is: {kb_result[0][0].value} ")
    print(f"Review subcollection count for App Document {app['id']} is: {r_result[0][0].value} ")
    print(f"Pricing Plans subcollection count for App Document {app['id']} is: {pp_result[0][0].value} ")
```

##### This is the code to get the count of all collections/subcollections in the Category top-level collection. The subcollection here is Apps.

```python
from google.cloud import firestore
from google.cloud.firestore_v1.base_query import FieldFilter
from google.cloud.firestore_v1 import aggregation

db = firestore.Client()
cat_ref = db.collection('Category')
query = cat_ref.count()
result = query.get()
print(f'Category collection has: {result[0][0].value} documents')

all_cat = cat_ref.stream()

for cat in all_cat:
    cat = cat.to_dict()
    app_ref = cat_ref.document(cat['id']).collection('Apps').count()
    app_result = app_ref.get()
    
    
    print(f"Apps subcollection count for Review Document {cat['id']} is: {app_result[0][0].value} ")
```

#### Top 10 Productivity Apps

```python
from google.cloud import firestore
from google.cloud.firestore_v1.base_query import FieldFilter

db = firestore.Client()
cat_ref = db.collection('Category')
cat_query = cat_ref.where(filter=FieldFilter('title','==','Productivity'))
cat_id = cat_query.get()[0].to_dict()
cat_id = cat_id['id']
app_ref = db.collection('Category').document(cat_id).collection('Apps')
query = app_ref.order_by('rating', direction=firestore.Query.DESCENDING).limit(10)
results = query.get()

for doc in results:
    entry = doc.to_dict()
    print(f'id: {entry["id"]}, title: {entry["title"]}, developer: {entry["developer"]}, rating: {entry["rating"]} , reviews_count: {entry["review_count"]}')

```

#### 10 Apps with Highest Number of Reviews

```python
from google.cloud import firestore
from google.cloud.firestore_v1.base_query import FieldFilter

db = firestore.Client()
app_ref = db.collection('Apps')
query = app_ref.order_by('review_count', direction=firestore.Query.DESCENDING).limit(10)
results = query.stream()

for doc in results:
    entry = doc.to_dict()
    print(f"id: {entry['id']}, title: {entry['title']}, developer: {entry['developer']}, rating: {entry['rating']} , reviews_count: {entry['review_count']}")
```
